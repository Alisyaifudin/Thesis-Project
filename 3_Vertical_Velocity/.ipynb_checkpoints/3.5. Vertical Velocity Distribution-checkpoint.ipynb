```python
data = data_raw.filter("z < 180").filter("z > 170").extract()
data
```

```python
from scipy.stats import gaussian_kde
jk = data["JK"].to_numpy()
w = data["w"].to_numpy()
djk = (jk.max()-jk.min())
dw = (w.max()-w.min())
wm = 0.05
jkm = 0.1

jk_grid, w_grid = np.mgrid[jk.min()-djk*jkm:jk.max()+djk*jkm:101j, w.min()-dw*wm:w.max()+dw*wm:100j]
positions = np.vstack([jk_grid.ravel(), w_grid.ravel()])
values = np.vstack([jk, w])
kernel = gaussian_kde(values)
dummy_kernel = np.rot90(kernel(positions).reshape(jk_grid.shape))
dummy_kernel.shape
```

```python
jks = jk_grid[:, 0]
ws = w_grid[0]
Djk = jks[1] - jks[0]
Dw = ws[1] - ws[0]
jks.shape
```

```python
# plot the results

plt.figure(figsize=(10,5))

plt.imshow(dummy_kernel, cmap=plt.cm.gist_earth_r, extent=[jk.min()-djk*jkm, jk.max()+djk*jkm, w.min()-dw*wm, w.max()+dw*wm], aspect='auto')
plt.colorbar()
plt.plot(jk, w, 'k.', markersize=0.5)
plt.xlabel("J-Ks")
plt.ylabel("w")
# plt.xlim(-0.1, 0.2)
# plt.ylim(np.min(mj)*0.9, np.max(mj)*1.1)
plt.gca().invert_yaxis()
plt.show()
```

```python
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={"projection": "3d"})

w_grid_r = w_grid[:,::-1]
# Plot the surface.
surf = ax.plot_surface(w_grid_r, jk_grid , dummy_kernel.T, cmap=plt.cm.gist_earth_r)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter('{x:.02f}')
ax.view_init(elev=80., azim=-30)
ax.set_ylabel("jk")
ax.set_xlabel("w")
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
from scipy.integrate import simps
index = 100 # 0-100
stds = []
jk_s = []
a_s = []
plt.figure(figsize=(10,5))
for index in range(20, 80, 2):
    j_k = jks[index] 
    jk_s.append(j_k)
    fw = dummy_kernel[::-1, index].copy()
    maks = np.max(fw)
    a_s.append(maks)
    area = simps(fw, ws)
    fw /= area
    mean = np.sum(fw*ws)/np.sum(fw)
    std = np.sqrt(np.sum(fw*(ws-mean)**2)/np.sum(fw))
    stds.append(std)
    print(f"mean: {mean:.2f}, std: {std:.2f}")
    
    plt.plot(ws, fw)

plt.show()
```

```python
def parabola(x, a, b, c):
    return a*x**2 + b*x + c

def line(x, a, b):
    return a*x + b

popt, pcov = curve_fit(line, jk_s, stds)
print(popt)
plt.scatter(jk_s, stds)
plt.plot(jk_s, line(np.array(jk_s), *popt), "r-")
```

```python
popt, pcov = curve_fit(parabola, jk_s, np.log(a_s))
print(popt)
plt.scatter(jk_s, np.log(a_s))
plt.plot(jk_s, parabola(np.array(jk_s), *popt), "r-")
```

```python
from scipy.optimize import curve_fit
from scipy.stats import norm

def gaussian(x, mu, sigma, A):
    return A*norm.pdf(x, mu, sigma)

ze = np.linspace(-200, 200, 15)
zs = (ze[1:] + ze[:-1])/2
JK_lower = 0.1
JK_upper = 0.2
JK_min = 0.08
JK_max = 0.3
JK = np.linspace(JK_min, JK_max, 2)
ww = np.linspace(-40, 40, 100)

dat = []

for JK_lower, JK_upper in zip(JK[:-1], JK[1:]):
  plt.figure(figsize=(7, 4))
  jk = (JK_lower + JK_upper)/2
  obj = {"JK": jk, "JK_lower": JK_lower, "JK_upper": JK_upper, "arr": []}
  for z_low, z_high in zip(ze[:-1], ze[1:]):
    mask = ((data_raw['JK'] > JK_lower)*(data_raw['JK'] < JK_upper)*(data_raw['z'] > z_low)*(data_raw['z'] < z_high))
    df_tes = data_raw[mask].extract()
    w = df_tes["w"].to_numpy()
    wnum, wedge = np.histogram(w, bins=10)
    ws = wedge[:-1] + (wedge[1]-wedge[0])/2
    p0 = [np.mean(w), np.std(w), np.sum(w)]
    popt, pcov = curve_fit(gaussian, ws, wnum, p0=p0)
    # df_tes.viz.histogram("w", shape=10, label=f"{z_low} < z < {z_high}", n="normalize")
    ls = "solid" if z_low < 0 else "dashed"
    print(z_low, z_high, len(df_tes), popt[0], popt[1])
    zz = (z_low + z_high)/2
    obj["arr"].append({"z": zz, "z_low": z_low, "z_high": z_high, "w_mean": popt[0], "w_std": popt[1], 
                      "A": popt[2], "w_num": len(df_tes)})
    plt.plot(ww, norm.pdf(ww, loc=0, scale=popt[1]), label=f"{np.round(z_low)} < z < {np.round(z_high)}", ls=ls)
  dat.append(obj)
  plt.legend()
  plt.title(f"{JK_lower} < J-K < {JK_upper}")
  plt.show()

print(zs, JK)
```

```python
def parabola(x, a, x0, y0):
    return a*(x-x0)**2 + y0

def line(x, a, b):
    return a*x + b

zz = np.linspace(-200, 200, 100)
plt.figure(figsize=(7, 4))
popts = []
jks = []
for d in dat:
  std = []
  z = []
  for a in d["arr"]:
    std.append(a["w_std"])
    z.append(a["z"])
  popt, pcov = curve_fit(parabola, z, std)
  plt.scatter(z, std, label=f"{d['JK_lower']} < J-K < {d['JK_upper']}")
  plt.plot(zz, parabola(zz, *popt), "r-")
  print(popt)
  popts.append(popt)
  jks.append(d["JK"])
plt.title(f"{d['JK_lower']} < J-K < {d['JK_upper']}")
plt.xlabel("z")
plt.ylabel("w_std")
plt.legend()
plt.show()
```

### Conclusion

\begin{equation}
  F(z, w, c) = N(w, z, c)\cdot \frac{\nu^*(z)}{\text{comp}(z)}
\end{equation}
where
\begin{equation}
  N(w, z, c) = \frac{1}{\sqrt{2\pi}\sigma_w(z, c)}\exp{\left[-\left(\frac{w-w0}{2\sigma_w(z, c)}\right)^2\right]}
\end{equation}
and 
\begin{equation}
  \sigma_w(z, c) = a(c)(z-z_0)^2 + \sigma_{w0}(c)
\end{equation}
with
\begin{aligned}
  a(c) = m_a c + a_0 \\
  \sigma_{w0}(c) = m_{\sigma} c + \sigma_0
\end{aligned}
also
\begin{equation}
  \nu^*(z) = \int \text{d}\nu^*(z, c) = \nu_0\int \text{d}A(c) \int \frac{N(w, z_0, c)\text{ d}w}{\sqrt{w^2-\Phi(z)}}
\end{equation}
where
\begin{equation}
  A(c) = e^{bc}
\end{equation}

```python
w_edges = np.linspace(-40,40,41)
ws = 0.5*(w_edges[1:]+w_edges[:-1])
w_a = a_type.w.to_numpy()
w_f = f_type.w.to_numpy()
w_g = g_type.w.to_numpy()
w_num_a, _ = np.histogram(w_a, w_edges)
w_num_f, _ = np.histogram(w_f, w_edges)
w_num_g, _ = np.histogram(w_g, w_edges)

ef_a = np.sqrt(w_num_a)
ef_f = np.sqrt(w_num_f)
ef_g = np.sqrt(w_num_g)
```

```python
plt.errorbar(ws, w_num_a/w_num_a.sum(), yerr=ef_a/w_num_a.sum(), fmt="o", label="A")
plt.errorbar(ws, w_num_f/w_num_f.sum(), yerr=ef_f/w_num_f.sum(), fmt="o", label="F")
plt.errorbar(ws, w_num_g/w_num_g.sum(), yerr=ef_g/w_num_g.sum(), fmt="o", label="G")
```
