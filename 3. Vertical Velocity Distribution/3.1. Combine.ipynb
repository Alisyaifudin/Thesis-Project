```python
import vaex
from glob import glob
import pandas as pd
import numpy as np
from os.path import join, abspath
from os import pardir, mkdir
```

```python
name = "Combine"
root_data_dir = abspath(join(pardir, "Data"))
data_dir = join(root_data_dir, name)
try:
  mkdir(data_dir)
  print(f"Creating {name} dir in Data dir")
except FileExistsError:
  print("Directory already exist. Good to go!")
data_dir
```

```python
# load Gaia-2MASS
name = "Gaia-2MASS"
root_data_dir = abspath(join(pardir, "Data"))
gaia_dir = join(root_data_dir, name)
gaia_files = glob(join(gaia_dir, "*.hdf5"))
gaia_files.sort()
gaia_files[:5]
```

# RAVE6

```python
name = "rave"
data_dir_rave = join(data_dir, name)
try:
  mkdir(data_dir_rave)
  print(f"Creating {name} dir in Data/Combine dir")
except FileExistsError:
  print("Directory already exist. Good to go!")
data_dir_rave
```

```python
name = "RAVE6"
rave_dir = join(root_data_dir, name)
rave_files = glob(join(rave_dir, "*.hdf5"))
rave_files.sort()
rave = vaex.open_many(rave_files)
rave
```

```python
# Join
cols = rave.column_names
for file in gaia_files:
    gaia = vaex.open(file)
    join_rave = gaia.join(rave, on="source_id", how="left")
    for col in cols:
        if col == "source_id": continue
        join_rave[col] = join_rave[col].to_pandas_series().to_numpy() # Do this to make np.nan value to no value
    join_rave.export(join(data_dir_rave, file[-17:]), progress=True) # save the result if necessary
```

```python
# sneak peek
vaex.open_many(glob(join(data_dir_rave, "*.hdf5")))
```
